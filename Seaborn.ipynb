### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot. 

**Ans:** Types of plot that can be plot using seaborn library

- Scatterplot : Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another. 
- Lineplot : Line plots are used to determine the relation between two datasets.
- Histogram : It is used to summarize discrete or continuous data that are measured on an interval scale
- Barplot: It allows you to compare different sets of data among different groups easily. 
- Boxplot : Box plots provide a quick visual summary of the variability of values in a dataset. 

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, style= fmri.event, hue= fmri.region)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.age)
```

```python
sns.barplot(x = titanic.pclass, y = titanic.fare)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds
```

```python
diamonds['cut'].value_counts()
```

```python
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', kde=True)
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
sns.pairplot(iris, hue= 'species')
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap. 

```python
flights = sns.load_dataset("flights")
flights
```

```python
# Pivot the data to create a matrix for heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(data=flights_pivot, annot=True)
```
