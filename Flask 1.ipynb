#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

**Ans:**
Flask is a popular micro web framework written in Python. It is designed to be lightweight and easy to use, making it an excellent choice for building web applications, APIs, and smaller projects.

Here are some advantages of Flask:

1. Simplicity: Flask follows a minimalist approach, providing only the essential tools to get started with web development. It has a small and intuitive core that is easy to understand and learn. This simplicity allows developers to quickly build applications without unnecessary overhead.

2. Flexibility: Flask provides a flexible framework that allows developers to choose the components they need for their specific requirements. It does not impose any specific project structure or dependencies, giving developers the freedom to make decisions based on their preferences or project constraints.

3. Lightweight: Flask is lightweight and has a small footprint. It does not require a large amount of code or dependencies, making it suitable for projects with limited resources. This lightweight nature also contributes to faster response times and improved performance.

4. Extensibility: Flask is highly extensible and provides a wide range of extensions and libraries that can be easily integrated into your project. These extensions cover various functionalities such as database integration, form validation, authentication, and more. The Flask community actively maintains and develops these extensions, ensuring a robust ecosystem.

5. Integrated Development Server: Flask comes with a built-in development server, allowing developers to quickly test their applications during the development process without the need for additional setup. This makes it convenient for rapid prototyping and debugging.

6. Scalability: While Flask is considered a micro framework, it is still capable of handling larger and more complex applications. It can be combined with other libraries or frameworks to add scalability and support for heavy traffic loads.

 #### Q3. What is App routing in Flask? Why do we use app routes?

**Ans**
In Flask, app routing refers to the process of defining URL routes or endpoints for handling different HTTP requests. Each route specifies a unique URL path and associates it with a Python function that will be executed when that specific URL is accessed.

App routes in Flask are used to define URL mappings and handle different HTTP requests within a web application. They allow developers to specify how the application responds to specific URLs and HTTP methods, enabling the implementation of different functionalities, such as displaying different pages, processing form submissions, or handling API endpoints. By organizing code logic based on routes, app routes promote modularity and maintainability. They also enable the construction of RESTful APIs by associating routes with specific resources and HTTP methods. Additionally, app routes facilitate URL building, ensuring consistent and reliable generation of URLs throughout the application. In summary, app routes are a crucial component of Flask that enables flexible request handling and facilitates the development of web applications and APIs.

##### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
- Company Name: ABC Corporation
- Location: India
- Contact Detail: 999-999-9999
**Attach the screenshot of the output in Jupyter Notebook.**

```python
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation\n"

@app.route('/')
def details():
    return "Company Name : ABC Corporation Location: India Contact Detail: 999-999-9999"

if __name__ == "__main__":
    app.run(host = "0.0.0.0")
```
